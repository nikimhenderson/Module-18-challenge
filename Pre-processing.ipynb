```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import warnings

# Hiding warnings
warnings.filterwarnings('ignore')

# File to Load
january_data = Path('Resources/JC-202401-citibike-tripdata.csv')
february_data = Path('Resources/JC-202402-citibike-tripdata.csv')
march_data = Path('Resources/JC-202403-citibike-tripdata.csv')
april_data = Path('Resources/JC-202404-citibike-tripdata.csv')
may_data = Path('Resources/JC-202405-citibike-tripdata.csv')

# Read Data File and store into Pandas DataFrames
january_df = pd.read_csv(january_data)
february_df = pd.read_csv(february_data)
march_df = pd.read_csv(march_data)
april_df = pd.read_csv(april_data)
may_df = pd.read_csv(may_data)
```

```python
#concatenate the data
concat_df = pd.concat([march_df, april_df, may_df],axis = 0)
```

```python
concat_df
```

```python
nan_count = concat_df.isna().sum()
print(nan_count)
```

```python
final_df = concat_df.dropna(subset=['start_station_name','end_station_name','start_lat','start_lng','end_lat','end_lng','start_station_id','end_station_id'])
final_df
```

```python
nan_count = final_df.isna().sum()
print(nan_count)
```

```python
final_df.info()
```

```python
station_coords_start = final_df.groupby('start_station_name').agg({
    'start_lat': 'mean',
    'start_lng': 'mean'
}).reset_index()
station_coords_start.value_counts()
```

```python
final_agg_df_start = pd.merge(station_coords_start, final_df, on='start_station_name', suffixes=('','_avg'))
final_agg_df_start = final_agg_df_start.rename(columns={'start_lat_avg': 'avg_start_lat', 'start_lng_avg': 'avg_start_lng'})
final_agg_df_start
```

```python
# Export out concatenated data
output_path = Path('Resources/final_data.csv')
final_df.to_csv(output_path, index=False)
```
